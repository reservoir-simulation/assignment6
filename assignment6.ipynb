$$
\mathbf{A}^\intercal \mathbf{A} \vec{x} = \mathbf{A}^\intercal \vec{b},
$$

which has the solution

$$
\vec{x} =\left(\mathbf{A}^\intercal \mathbf{A}\right)^{-1} \mathbf{A}^\intercal \vec{b}.
$$

$$
\mathbf{A} =
\begin{bmatrix}
1 & f(\phi_1) \\
1 & f(\phi_2) \\
\vdots & \vdots \\
1 & f(\phi_N)
\end{bmatrix},
$$

$$

\begin{matrix}
\kappa_1 \\
\kappa_2 \\
\vdots \\
\kappa_N
\end{matrix}
\right\rbrace
$$
